```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pyarrow.parquet as pq
import pyarrow as pa
```

```python
filename = '/data/cewalden/soc16.parquet'
data = pq.read_table(filename).to_pandas()
data.head()
```

```python
dfStr = data.copy().astype(str)
```

```python
qCols = ['AREA', 'FNSQ', 'FFNSQ', 'LOTV', 'PVALU', 'FSQFS', 'CONPR', 'SLPR', 'SQFS', 'WEIGHT', 'FCONPR', 'FSLPR']
for i in qCols:
    dfStr.drop(i, axis = 1)
    dfStr[i] = data[i]
    
dateCols = ['COMP_D', 'SALE_D', 'STRT_D', 'AUTH_D']
for i in dateCols:
    dfStr.drop(i, axis = 1)
    dfStr[i] = data[i]
    
dCols = ['COMP', 'SALE', 'STRT', 'AUTH']
for i in dCols:
    dfStr.drop(i, axis = 1)
    dfStr[i] = data[i]

cCols = ['ACS', 'AGER', 'ASSOC', 'BASE', 'CAT', 'CON', 'DECK', 'DET', 'DIV',
       'FINC', 'FNBS', 'FOYER', 'FRAME', 'GAR', 'HEAT', 'HEAT2', 'LNDR',
       'METRO', 'MFGS', 'PATI', 'PRCH', 'SEWER', 'STOR', 'WAL1', 'WAL2',
       'WALS', 'WATER', 'BEDR', 'FPLS', 'FULB', 'HAFB', 'FUEL', 'FUEL2']

flags = ['AREA_F', 'FNSQ_F', 'SLPR_F', 'FSLPR_F', 
         'CONPR_F', 'FCONPR_F', 'LOTV_F', 
         'SQFS_F', 'FSQFS_F', 'PVALU_F', 'FFNSQ_F']
```

```python
colors =  ['#000000', '#800000', '#008000',  '#808000', '#000080',
            '#800080', '#008080', '#c0c0c0', '#808080', '#ff0000', 
            '#00ff00', '#ffff00', '#0000ff', '#ff00ff', '#00ffff', 
            '#000000', '#00005f', '#005f5f', '#005faf', '#005fd7', 
            '#008700', '#00875f', '#00afff', '#00d700', '#00d787', 
            '#00d7ff', '#00ff00', '#00ffff', '#5f0000', 
            '#5f005f', '#5f00af', '#5f5f00', '#5f5f5f', '#5f5faf', 
            '#5f5fff', '#5f8700', '#5f875f', '#5f87af', '#5f87ff', 
            '#5faf00', '#5faf5f', '#5fafaf', '#5fafff', '#5fd700', 
            '#5fd75f', '#5fd7af']

divLabels = ['New England', 'Middle Atlantic', 'East North Central', 
            'West North Central', 'South Atlantic', 'East South Central',
            'West South Central', 'Mountain', 'Pacific']
```

Getting an idea of the distribution of values in categorical columns

```python
cols = cCols
fig, axes = plt.subplots(len(cols), 1, figsize = (5,30))
for i, ax in enumerate(axes):
    binS = 50
    if len(dfStr[cols[i]].unique()) < 50:
        binS = len(dfStr[cols[i]].unique())
    ax.hist(dfStr[cols[i]], color= colors[i], weights = dfStr['WEIGHT'], bins = binS)

plt.legend(
    [plt.Line2D((0,1),(0,0), color=colors[i]) for i in range(len(cCols))],
    cCols, bbox_to_anchor=(1.5, 2*len(cols)))
plt.subplots_adjust(hspace=1)
plt.show()
```

Visualizing the distribution of values in the date columns

```python
cols = dCols
fig, axes = plt.subplots(len(cols), 1, figsize = (10,15))
for i, ax in enumerate(axes):
    binS = 50
    if len(dfStr[cols[i]].unique()) < 50:
        binS = len(dfStr[cols[i]].unique())
    ax.hist(dfStr[dfStr[cols[i]] != 0][cols[i]], color= colors[i], weights = dfStr[dfStr[cols[i]] != 0]['WEIGHT'], bins = binS)

plt.legend(
    [plt.Line2D((0,1),(0,0), color=colors[i]) for i in range(len(cols))],
    cols, bbox_to_anchor=(1.5, 2*(len(cols)-1)))
plt.subplots_adjust(hspace=1)
plt.show()
```

```python
def plotGroupsHist(group, column, sizeX = 5, sizeY = 5, df = dfStr):
    """Makes a series of graphs using the group parameter as groups
    
    Parameters
    ----------
    group: String
        Name of column to use for grouping
    column: String
        Name of the column to be displayed as a histogram
        
    Optional Parameters
    -------------------
    sizeX: real number
        How wide the graphs should be
    sizeY: real number
        How tall the overall graph should be"""
    cats = np.sort(df[group].unique())
    fig, axes = plt.subplots(len(cats), 1, figsize = (sizeX, sizeY))
    plt.title(column + " grouped by " + group, y = 2*len(cats) - 1)
    binS = 50
    if len(df[column].unique()) < 50:
        binS = len(df[column].unique())
    for i, ax in enumerate(axes):
        ax.hist(df[df[group] == cats[i]][column], 
                weights = df[df[group] == cats[i]]['WEIGHT'],
                color = colors[i], 
                bins = binS)
    if group == 'DIV':
        labels = divLabels
    else:
        labels = cats
    plt.legend(
            [plt.Line2D((0,1),(0,0), color=colors[i], lw = 7) for i in range(len(cats))],
            labels, bbox_to_anchor=(1.5, 2*(len(cats)-1)))
    plt.subplots_adjust(hspace=1)
    plt.show()
```

```python
plotGroupsHist('DIV', 'WAL1', sizeY = 7)
```

The primary exterior wall material varies interestingly with the division. Vinyl siding is the by far the most popular material popular in the regions of New England, Middle Atlantic, and East and West North Central. It is also the most popular primary exterior wall material in the United States.

However, the other regions prefer materials. South Atlantic built houses with a East and West South Central preferred brick or brick veneer over vinyl, and Mountain and Pacific preferred stucco with fiber cement siding coming in second.

```python
plotGroupsHist('DIV', 'BEDR', sizeY = 10)
```

Most single-family all over the United States are built with 3 bedrooms. However, the Pacific region builds a majority of its homes with 4 bedrooms.

```python
plotGroupsHist('DIV', 'SLPR', sizeY = 10, df = dfStr[dfStr['SLPR'] > 0])
```

```python
plotGroupsHist('DIV', 'COMP', sizeY = 10, df = dfStr[dfStr['COMP'] > 201399])
```

```python
plotGroupsHist('BASE', 'PRCH', sizeY = 10)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

cat = 'GAR'
y = dfStr[cat].values
X = dfStr.drop(dateCols, axis = 1).drop(cat, axis = 1).values

reg = LinearRegression()
cv_scores = cross_val_score(reg, X, y, cv=5)
# Print the 5-fold cross-validation scores
print(cv_scores)

print("Average 5-Fold CV Score: {}".format(np.mean(cv_scores)))
```
