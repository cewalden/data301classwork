## Hat potential

The following potential is often used in Physics and other fields to describe symmetry breaking and is often known as the "hat potential":

$$ V(x) = -a x^2 + b x^4 $$

Write a function `hat(x,a,b)` that returns the value of this function:

```python
def hat(x, a, b):
    return b * (x ** 4) - a * (x ** 2)
```

```python
assert hat(0.0, 1.0, 1.0)==0.0
assert hat(0.0, 1.0, 1.0)==0.0
assert hat(1.0, 10.0, 1.0)==-9.0
```

Plot this function over the range $x\in\left[-3,3\right]$ with $b=1.0$ and $a=5.0$:

```python
a = 5.0
b = 1.0
```

```python
arr = np.arange(-3,4)
data1 = hat(arr, a, b)
plt.plot(arr, data1)
plt.grid(axis = 'both')
plt.show()
```

Write code that finds the two local minima of this function for $b=1.0$ and $a=5.0$.

* Use `scipy.optimize.minimize` to find the minima. You will have to think carefully about how to get this function to find both minima.
* Print the x values of the minima.

```python
dataTable = np.zeros(0)
for i in arr:
    dataTable= np.append(dataTable, opt.minimize(lambda theta: hat(i, theta[0], theta[1]), [a,b]))
minA = 100000
minB = 100000
for i in range(len(dataTable)):
    temp = dataTable[i].x
    if hat(temp[0], a, b) < hat(minA, a, b):
        minA = temp[0]
    if hat(temp[1], a, b) < hat(minB, a, b):
        minB= temp[1]
minA -= 3
minB -= 3
print(minB, minA)

plt.plot(arr, data1)
plt.scatter([minB, minA], [hat(minB, a, b), hat(minA, a, b)], c = 'r')
plt.grid(axis = 'both')
plt.title('Hat potential of values, with minumum')
plt.xlabel('Value')
plt.ylabel('Hat potential of value')
plt.show()
```
